```python
import qMS
import vizLib
import qMSDefs
import pandas
import numpy
import matplotlib.pyplot as plt
import pylab
import glob
import scipy
from scipy.interpolate import interp1d
```

```python
path = '/home/jhdavis/JStokes/2014_01_15-JStokesSucroseGradientProfiles/'

files = glob.glob(path+'*.txt')
vizLib.setRcs(scale=12, legendScale=16)
dat = pandas.read_table(files[0],names=['time', 'wt', '1 hour', '6 hour'], index_col=False, header=1)
```

```python
tnew = numpy.linspace(2, 69, 1000)
f2wt = interp1d(dat['time'], dat['wt'], kind='cubic')
f2hr1 = interp1d(dat['time'], dat['1 hour'], kind='cubic')
f2hr6 = interp1d(dat['time'], dat['6 hour'], kind='cubic')
```

```python
wtOffset=0.0
hr1Offset=0.5
hr6Offset=1.0
Red = pylab.cm.Reds(0.5)
Blue = pylab.cm.Blues(0.5)
Green = pylab.cm.Greens(0.5)
lw = 6
f = pylab.figure(figsize=(6,4))
ax = f.add_subplot(111)

#ax.plot(dat['time'], dat['wt'], label='wt', lw=lw, color=Red)
ax.plot(tnew, f2wt(tnew), label='wt', lw=lw, color=Red)
ax.plot(tnew, f2hr1(tnew)-10, label='1hr', lw=lw, color=Blue)
ax.plot(tnew, f2hr6(tnew)-13, label='6hr', lw=lw, color=Green)

ax.set_xlim((32, 45))
ax.set_ylim((0, 100))
ax.set_xticks([])
ax.set_xticks([32,45])
ax.set_yticks([])
#ax.set_yticks([0, 25, 50, 100, 200])
vizLib.cleanAxis(ax, complete=True)

pylab.legend()
pylab.tight_layout()

#pylab.savefig(path+'30Sprofiles.pdf')
#pylab.savefig('/home/jhdavis/Dropbox/fromLinux/gradientProfiles.pdf')
```
